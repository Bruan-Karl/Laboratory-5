```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
K = np.array([9, 12, 3])
L = np.array([11,0,6])

fig = plt.figure()
ax1 = fig.gca(projection='3d')

ax1.set_xlim([-4, 15])
ax1.set_ylim([-4, 15])
ax1.set_zlim([-4, 15])

#origin = (0,0)
ax1.quiver(0, 0, 0, K[0], K[1], K[2], arrow_length_ratio=0.05, colors='r')
ax1.quiver(0,0,0, L[0], L[1], L[2], arrow_length_ratio=0.05, colors='b')


plt.show()
```
